Criando Mini-Batch Gradient Descent

```python
#https://www.geeksforgeeks.org/ml-mini-batch-gradient-descent-with-python/

def create_mini_batches(X, y, batch_size): 
    mini_batches = [] 
    data = np.hstack((X, y)) 
    np.random.shuffle(data) 
    n_minibatches = data.shape[0] // batch_size 
    i = 0
  
    for i in range(n_minibatches + 1): 
        mini_batch = data[i * batch_size:(i + 1)*batch_size, :] 
        X_mini = mini_batch[:, :-1] 
        Y_mini = mini_batch[:, -1].reshape((-1, 1)) 
        mini_batches.append((X_mini, Y_mini)) 
    if data.shape[0] % batch_size != 0: 
        mini_batch = data[i * batch_size:data.shape[0]] 
        X_mini = mini_batch[:, :-1] 
        Y_mini = mini_batch[:, -1].reshape((-1, 1)) 
        mini_batches.append((X_mini, Y_mini)) 
    return mini_batches 
```

Definindo a função GD

```python
def gradiente_descend_step(w0, w1, X, y, alpha):
    erro_w0 = 0
    erro_w1 = 0
    n=float(len(X))
    for i in range (0, len(X)):
        erro_w0 = y_pred(X[i],w0,w1)-y[i]
        erro_w1 = (y_pred(X[i],w0,w1)-y[i])* X[i]
    
    new_w0 = w0 - alpha * (1/n) * erro_w0
    new_w1 = w1 - alpha * (1/n) * erro_w1
    
    return new_w0, new_w1
```

LOOP ALGORITMO GRADIENTE DESCENDENTE

```python
epoch=1000
```

```python
def gradiente_descend(w0, w1, X, y, alpha, epoch, batch_size=4):
    custo=np.zeros(epoch)
    for i in range (epoch):
        mini_batches = create_mini_batches(X, y, batch_size)
        for mini_batch in mini_batches:
            X_mini, y_mini = mini_batch
            if len(X_mini) == 0:
                break
            w0, w1 = gradiente_descend_step(w0, w1, X_mini, y_mini, alpha)             
            custo[i]=custo_MSE(X_mini,y_mini,w0,w1)
    return w0,w1,custo
```

```python
w0, w1, custo = gradiente_descend(w0, w1, X, y, alpha, epoch, batch_size=4)
```
