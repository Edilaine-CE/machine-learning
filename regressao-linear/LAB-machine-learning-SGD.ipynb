### LAB 1 - ALGORITMO GRADIENTE DESCENDENTE ESTOCÁSTICO - REGRESSÃO LINEAR SIMPLES

```python
#importacoes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

DADOS

```python
data_table = pd.DataFrame({'Altura': np.array([42.8, 63.5, 37.5, 39.5, 45.5, 38.5, 43.0, 22.5, 37.0, 23.5, 33.0, 58.0]) / 39.37,
                            'Cateter': [37, 50, 34, 36, 43, 28, 37, 20, 34, 30, 38, 47]},
                            dtype='float')
```

```python
X=data_table[['Altura']].values
y=data_table[['Cateter']].values
```

TAXA DE APRENDIZADO (ALPHA) - aceleração da descida

```python
alpha = 0.01
```

VALORES INICIAIS w0 e w1 

```python
w0 = 0.1
w1=0.1
```

DEFININDO MODELO DE PREDIÇÃO

```python
def y_pred (x, w0, w1):
    return w0 + w1*x
```

FUNÇÃO CUSTO LMS (Least Mean Squares) 

```python
#def custo_LMS(X,y,w0,w1):
#    erro=0
#    for i in range (0, len(X)):
#        erro+=(y_pred(X[i],w0,w1)-y[i])
#    return erro

def custo_MSE(X,y,w0,w1):
    erro=0
    n=float(len(X))
    for i in range (0, len(X)):
        erro+=(y_pred(X[i],w0,w1)-y[i])**2
    return erro/n
```

ALGORITMO GRADIENTE DESCENDENTE ESTOCÁSTICO 

Definindo a função SGD

```python
def gradiente_descend_step(w0, w1, X, y, alpha):
    erro_w0 = 0
    erro_w1 = 0
    for i in range (0, len(X)):
        erro_w0 = y_pred(X[i],w0,w1)-y[i]
        erro_w1 = (y_pred(X[i],w0,w1)-y[i])* X[i]
        new_w0 = w0 - alpha * erro_w0
        new_w1 = w1 - alpha * erro_w1
    
    return new_w0, new_w1
```

LOOP ALGORITMO GRADIENTE DESCENDENTE

```python
epoch=150
```

```python
def gradiente_descend_stocastic(w0, w1, X, y, alpha, epoch):
    custo=np.zeros(epoch)
    for i in range (epoch):
        w0, w1 = gradiente_descend_step(w0, w1, X, y, alpha)
        custo[i]=custo_MSE(X,y,w0,w1)
        np.random.shuffle(X)
    return w0,w1,custo
```

```python
w0, w1, custo = gradiente_descend_stocastic(w0, w1, X, y, alpha, epoch)
```

PLOTANDO O CUSTO

```python
fig, ax = plt.subplots()
ax.plot(np.arange(epoch), custo, 'r')
ax.set_xlabel('Iterações')
ax.set_ylabel('Custo')
ax.set_title('LMS vs. Epoch')
```

PLOTANDO HIPÓTESE OTIMIZADA (PONTOS E REGRESSÃO)

```python
def plot_line(X, y, w0, w1):
    x_values = [i for i in range (int(min(X))-1,int(max(X))+2)]
    y_values = [y_pred(x, w0, w1) for x in x_values]
    plt.plot(x_values,y_values,'r')
    plt.plot(X,y,'bo') 
    plt.xlabel('Altura(m)')
    plt.ylabel('Comprimento do Cateter (cm)')
```

```python
plot_line(X, y, w0, w1)
```
